```python
import os
import cv2
import numpy as np

base_input_path = r"D:\LLM\DATASET\Segmentation"
base_output_path = r"D:\LLM\done\enhanced_overlays"
os.makedirs(base_output_path, exist_ok=True)

classes = ['Glioma', 'Meningioma', 'pituitary tumor']

def extract_index(filename):
    """Extracts the numeric index from filenames like 'image_10.png'"""
    try:
        return int(os.path.splitext(filename)[0].split('_')[-1])
    except ValueError:
        return -1  

for cls in classes:
    input_folder = os.path.join(base_input_path, cls)
    output_folder = os.path.join(base_output_path, cls)
    os.makedirs(output_folder, exist_ok=True)

    all_files = os.listdir(input_folder)
    valid_files = [f for f in all_files if extract_index(f) != -1]
    sorted_files = sorted(valid_files, key=extract_index)

    for i in range(0, len(sorted_files), 2):
        try:
            brain_path = os.path.join(input_folder, sorted_files[i])
            mask_path = os.path.join(input_folder, sorted_files[i + 1])

            brain = cv2.imread(brain_path)
            mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

            if brain is None or mask is None:
                print(f"Skipped: {brain_path}, {mask_path}")
                continue

            # Create a red mask overlay
            red_mask = np.zeros_like(brain)
            red_mask[:, :, 2] = mask

            # Blend overlay
            overlay = cv2.addWeighted(brain, 0.8, red_mask, 0.4, 0)

            out_path = os.path.join(output_folder, f'overlay_{i // 2}.png')
            cv2.imwrite(out_path, overlay)

        except IndexError:
            print(f"Unpaired file at index {i} in class {cls}, skipping.")

    print(f"[✓] Done processing {cls}")
```

```python
import torch
from open_clip import create_model_and_transforms, get_tokenizer
from PIL import Image

device = "cuda" if torch.cuda.is_available() else "cpu"

model, _, preprocess = create_model_and_transforms(
    'hf-hub:microsoft/BiomedCLIP-PubMedBERT_256-vit_base_patch16_224'
)
tokenizer = get_tokenizer(
    'hf-hub:microsoft/BiomedCLIP-PubMedBERT_256-vit_base_patch16_224'
)

model = model.to(device).eval()
```

```python
captions = [
    # Glioma
    "Segmented glioma tumor appears in the right temporal region.",
    "MRI scan shows glioma with mass effect.",
    "Glioma tumor with irregular margins in the frontal lobe.",
    
    # Meningioma
    "Segmented region corresponds to a meningioma near the frontal lobe.",
    "MRI suggests a meningioma with dural attachment.",
    "Meningioma adjacent to the falx cerebri seen on MRI.",
    
    # Pituitary Tumor
    "Segmented pituitary tumor located near optic chiasm.",
    "MRI shows pituitary tumor compressing surrounding structures.",
    "Pituitary tumor extending into the suprasellar cistern."
]
```

```python
import os
import random
import torch
import matplotlib.pyplot as plt
from PIL import Image

image_folder = r"D:\LLM\done\enhanced_overlays\Glioma"

image_paths = sorted([
    os.path.join(image_folder, f)
    for f in os.listdir(image_folder)
    if f.endswith(".png")
])

N = 5
sampled_images = random.sample(image_paths, min(N, len(image_paths)))

def generate_caption(image_path):
    image = preprocess(Image.open(image_path)).unsqueeze(0).to(device)
    text_tokens = tokenizer(captions).to(device)

    with torch.no_grad():
        image_features = model.encode_image(image)
        text_features = model.encode_text(text_tokens)

        image_features = image_features / image_features.norm(dim=-1, keepdim=True)
        text_features = text_features / text_features.norm(dim=-1, keepdim=True)

        similarity = (100.0 * image_features @ text_features.T)
        values, indices = similarity.softmax(dim=-1).max(dim=-1)

    return captions[indices.item()], values.item()


for image_path in sampled_images:
    caption, confidence = generate_caption(image_path)

    img = Image.open(image_path)
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"{caption}\nConfidence: {round(confidence, 4)}", fontsize=10)
    plt.show()
```
